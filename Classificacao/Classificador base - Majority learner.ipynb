## Aplicando o algoritmo

```python
base_credit = Orange.data.Table('base_credit_majority.csv')
```

#### Se fosse necessário dividir a base em treinamento e teste, para aplicação de outros métodos usando o pacote Orange3, teríamos:

### Aplicando Majority Learner

```python
majority = Orange.classification.MajorityLearner()
```

```python
previsoes = Orange.evaluation.testing.TestOnTestData(base_credit, base_credit,[majority])
```

```python
Orange.evaluation.CA(previsoes)
```
